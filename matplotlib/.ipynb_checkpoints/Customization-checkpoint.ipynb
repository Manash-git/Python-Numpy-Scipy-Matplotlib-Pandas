```python
import numpy as np
```

```python
t= np.array([1,2,3,4])
```

```python
plt.plot(t**2, color= 'red', label = 'square')
```

```python
plt.plot(t**3, color= 'green', label = 'cube')
```

```python
plt.plot(t**2, color= 'red', label = 'square')
plt.plot(t**3, color= 'green', label = 'cube')
```

```python
plt.plot(t**2, color= 'red', label = 'square')
plt.plot(t**3, color= 'green', label = 'cube')
plt.legend(["Square","Cube"])
```

```python
plt.plot(t**2, color= 'red', label = 'square')
plt.plot(t**3, color= 'green', label = 'cube')
plt.legend(["Square","Cube"], loc=(1,1))
```

```python
plt.plot(t**2, color= 'red', label = 'square')
plt.plot(t**3, color= 'green', label = 'cube')
plt.legend(["Square","Cube"], loc=(1,1))
plt.tight_layout
```

```python
plt.plot(t**2, color= 'red', label = 'square')
plt.plot(t**3, color= 'green', label = 'cube')
plt.legend(["Square","Cube"], loc=(1,1), facecolor = 'yellow', fontsize = 20, title= "My Legend")
plt.tight_layout()
```
