## Call signatures:

plot([x], y, [fmt], *, data=None, **kwargs)

plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)

```python
import matplotlib.pyplot as plt
```

```python
# red line 

plt.plot([10,20,30,40],'r')
```

```python
# red dot

plt.plot([10,20,30,40],'r+')
```

```python
# point size scaling 

plt.plot([10,20,30,40],'r+', markersize= 20)
```

```python
# plot line purutto size scaling 

plt.plot([10,20,30,40],'r', linewidth= 10)
```

```python
# two function plot in a single space
# [] - er moddhe all x axis borabor print hobe

plt.plot([10,20,30,40],'g',[5,10,20,39],'ro',linewidth=15, markersize= 10)
```

```python
import numpy as np
```

```python
a= np.arange(0,10,.5)
# it will plot according to y axis , because dont use [ ] 
plt.plot(a,'go')
```
