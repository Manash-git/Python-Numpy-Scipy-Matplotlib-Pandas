```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
degree = [0,30,45,60,90]
x = [ i* (np.pi / 180 ) for i in degree]
```

```python
plt.plot( degree, np.sin(x), marker= '+' , label='Sin')
plt.plot( degree, np.cos(x), marker= '^' , label='Cos')
```

```python
plt.plot( degree, np.sin(x), marker= '+' , label='Sin')
plt.plot( degree, np.cos(x), marker= '^' , label='Cos')
# plt.xticks(degree)
plt.title(" Sin and Cos Graph")
plt.xlabel('Angle')
plt.ylabel('Value')
plt.legend()
```
