```python
import pandas as pd
df = pd.read_csv("survey_results_public.csv")
# df = pd.read_csv("survey_results_public.csv", index_col= 'Respondent')
# df.head()
```

```python
country_response = df['Country'].value_counts()
```

```python
country_response
```

```python
country_grp = df.groupby(["Country"])
```

```python
country_use_Python= country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_use_Python
```

```python
new_DF = pd.concat([country_response, country_use_Python], axis = 'columns', sort = False)
new_DF
```

```python
new_DF.rename(columns={ 'Country': "Respondents",'LanguageWorkedWith':'Python Developers'}, inplace = True)
```

```python
new_DF
```

```python
new_DF['% (Python Developers/Respondents)'] = (new_DF['Python Developers']/new_DF['Respondents']) * 100
```

```python
new_DF.drop(columns='% (Respondents/Python Developers)',inplace= True)
```

```python
new_DF
```

```python
new_DF.sort_values(by= '% (Python Developers/Respondents)', ascending= False)
```
